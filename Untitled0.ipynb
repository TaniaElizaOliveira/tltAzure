```python
import requests
from bs4 import BeautifulSoup

def extract_text_from_url(url):
    response = requests.get(url)
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')

        # Remove script, style, and irrelevant tags
        for script_or_style in soup(['script', 'style', 'nav', 'footer', 'header', 'aside']):
            script_or_style.decompose()

        # Extract text and split into paragraphs
        raw_text = soup.get_text(separator=' ')

        # Clean and organize text into paragraphs
        lines = (line.strip() for line in raw_text.splitlines())
        paragraphs = [line for line in lines if line]  # Keep only non-empty lines

        # Join paragraphs with a double newline to mimic paragraph separation
        formatted_text = '\n\n'.join(paragraphs)
        return formatted_text
    else:
        print(f"Failed to fetch the URL. Status code: {response.status_code}")
        return None

# Chamar a função e imprimir o resultado
text = extract_text_from_url('https://dev.to/devops_den/azure-certification-for-developers-3jk0')
print(text)

```
